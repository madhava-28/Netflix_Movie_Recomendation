```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#loading the data
data=pd.read_excel('netflix recomendation.xlsx',header=None)
```

```python
data
```

```python
transaction=[]
for i in range(7466):
    transaction.append([str(data.values[i,j]) for j in range(20)])
transaction
```

```python
# most watched movie
m=dict()
for i in range(len(transaction)):
    for j in range(len(transaction[i])):
        if transaction[i][j] not  in m:
            m[transaction[i][j]]=1
        else:
            m[transaction[i][j]]+=1
m            
```

```python
import numpy as np
keys = list(m.keys())
values = list(m.values())
sorted_value_index = np.argsort(values)
sorted_value_index=np.flip(sorted_value_index)
sorted_dict = {keys[i]: values[i] for i in sorted_value_index}
 
print(sorted_dict)
```

# traning the model

```python
from apyori import apriori
basket=apriori(transactions=transaction,min_support=0.002,min_confidence=0.002,min_lift=3,min_length=2,max_length=2)
```

```python
res=list(basket)
```
